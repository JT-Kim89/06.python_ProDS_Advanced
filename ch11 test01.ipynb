### 3번

출퇴근 시간 배차간격 조정을 위해 우선적으로 각 정류소별 시간대별 출근 시간 승차 패턴을 파악하고자 한다. 지선 버스 노선을 대상으로 각 정규장별 승차인원을 계층적 군집분석을 활용하여 6개의 군집으로 분할하였을 때 출근시간대에 승차 인원이 가장 많은 정류소 군집의 번호는 몇 번인가? (정답 예시: 2)  

1) 출근 시간대는 오전 7시부터 9시까지로 정의  
2) 군집분석 시 자료는 Min-Max 정규화 실시  
3) 거리 계산은 유클리디안 거리로 하고 유사도는 Ward.D 방법을 사용  

```python
df_q3 = df_join.copy()
df_q3.head()
```

```python
df_q3 = df_q3.loc[df_q3["type"] == "지선", ]
df_q3.head()
```

```python
df_q3 = pd.concat([df_q3.loc[:, ["Year_Month", "Station_ID"]],
                  df_q3.loc[:, "H01": "H24"]],
                 axis = 1)
df_q3.head()
```

```python
df_q3_agg = df_q3.iloc[:, 1:].groupby("Station_ID").sum().reset_index()
df_q3_agg.head()
```

```python
df_q3_melt = df_q3_agg.melt(id_vars = "Station_ID")
df_q3_melt.head()
```

```python
df_q3_melt = df_q3_melt.sort_values(["Station_ID", "variable"], ascending = [False, True])
df_q3_melt.head()
```

```python
def nor_minmax(x):
    result = (x - x.min()) / (x.max() - x.min())
    return result
```

```python
df_q3_melt["value_nor"] = df_q3_melt.groupby("Station_ID")["value"].transform(nor_minmax)
df_q3_melt.head()
```

```python
df_q3_pivot = pd.pivot(df_q3_melt, index = "Station_ID", columns = "variable", values = "value_nor")
df_q3_pivot.head()
```

```python
model = AgglomerativeClustering(n_clusters = 6).fit(df_q3_pivot)
```

```python
df_q3_pivot["cluster"] = model.labels_
df_q3_pivot.head()
```

```python
df_q3_pivot = df_q3_pivot.reset_index()
df_q3_pivot.head()
```

```python
df_q3_pivot_melt = df_q3_pivot.melt(id_vars = ["Station_ID", "cluster"])
df_q3_pivot_melt.head()
```

```python
df_q3_pivot_melt["rush_hour"] = df_q3_pivot_melt["variable"].isin(["H07", "H08", "H09"]) + 0
df_q3_pivot_melt = df_q3_pivot_melt.loc[df_q3_pivot_melt["rush_hour"] == 1, ]
df_q3_pivot_melt = df_q3_pivot_melt.loc[df_q3_pivot_melt["value"] == 1, ]
df_q3_pivot_melt.head()
```

```python
len(df_q3_pivot_melt)
```
