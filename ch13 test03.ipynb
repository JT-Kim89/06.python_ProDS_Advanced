```python
df_q1 = df.copy()
df_q1.head()
```

```python
df_q1["Date"] = pd.to_datetime(df_q1["Date"])
df_q1["yesr"] = df_q1["Date"].dt.year
df_q1["count"] = 1
df_q1.head()
```

```python
df_q1_cnt = df_q1.groupby(["yesr", "Item"])["count"].sum().reset_index()
df_q1_cnt.head()
```

```python
df_q1_cnt_pivot = df_q1_cnt.pivot(index = "Item", columns = "yesr", values = "count").reset_index()
df_q1_cnt_pivot.head()
```

```python
df_q1_cnt_pivot = df_q1_cnt_pivot.fillna(0)
df_q1_cnt_pivot.head()
```
